# Handling Missgin vlaues

Drop company (94% missing values),the majority of its values are missing and also not crucial for the analysis.

```python
data_set.drop('company',  axis='columns', inplace = True)
```

```python
data_set.isnull().sum()
```

For the agent column, fill missing values with 0 (assuming no agent was used) and keep the column for analyzing booking channels.

```python
data_set['agent'].fillna(0, inplace=True)
```

```python
#convert agent coulmn data type as integer for consistency in data
data_set['agent'].astype(int) 
```

Converts the agent column to integer type for consistency, since agents are represented by numeric IDs.

```python
data_set.isnull().sum()
```

The country column, fill missing values with the mode (most frequent value)

We use "mode" for the country column because it's categorical data (countries). Mean and median are used for numerical data, not categories.

```python
#fill missing values in country column using mode
data_set['country'].fillna(data_set['country'].mode()[0], inplace=True)
```

```python
data_set.isnull().sum()
```

only 4 values are missing in the children column (0.003% of the data), it's not critical, but itâ€™s still best to handle them.

```python
#use median for fill missgin values in children column
data_set['children'].fillna(data_set['children'].median(), inplace=True)
```

```python
data_set.isnull().sum().sum()
```


- Agent: Filled missing values with 0.
- Country: Filled missing values with the most frequent value (mode).
- Children: Filled missing values with the median.
- All missing values are now handled, and the dataset is complete.

Visualize Booking Trends Over Time

```python
# visualise booking per month

# now first grouping bookings by month
booking_per_month = data_set.groupby('arrival_date_month')['is_canceled'].count()

plt.figure(figsize=(12,7)) 
sns.barplot(x=booking_per_month.index, y=booking_per_month.values, palette='viridis')  
# total number of bookings for each bar
for i, value in enumerate(booking_per_month.values):
    plt.text(i, value + 5, f'{value}', ha='center', fontsize=12, fontweight='bold', color='black') 

plt.title('Bookings Per Month', fontsize=18, fontweight='bold', color='darkblue', pad=20)
plt.xlabel('Months', fontsize=14)
plt.ylabel('Number Of Bookings', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
booking_per_month = data_set.groupby('arrival_date_month')['is_canceled'].count()

# Calculate total cancellations to compute percentages
total_booking = booking_per_month.sum()


plt.figure(figsize=(12,7)) 
sns.barplot(x=booking_per_month.index, y=booking_per_month.values, palette='plasma')

# Add percentage above each bar
for i, value in enumerate(booking_per_month.values):
    percentage = (value / total_booking) * 100
    plt.text(i, value + 5, f'{percentage:.2f}%', ha='center', fontsize=12, fontweight='bold', color='black')  


plt.title('Booking Per Month (In Percentage)', fontsize=18, fontweight='bold', color='darkblue', pad=20)
plt.xlabel('Months', fontsize=14)
plt.ylabel('Number Of Booking', fontsize=14)

plt.show()
```

This bar chart visualizes the number of bookings each month, allowing you to detect any seasonal booking trends.

```python
# Visualize Cancellations Per Month

# Frist, grouping cancellations by month
cancellation_per_month = data_set[data_set['is_canceled'] == 1].groupby('arrival_date_month')['is_canceled'].count()

plt.figure(figsize=(12,7)) 
sns.barplot(x=cancellation_per_month.index, y=cancellation_per_month.values, palette='plasma')  

# total number of cancellations for each bar
for i, value in enumerate(cancellation_per_month.values):
    plt.text(i, value + 5, f'{value}', ha='center', fontsize=12, fontweight='bold', color='black') 

plt.title('Cancellation Per Month', fontsize=18, fontweight='bold', color='darkblue', pad=20)
plt.xlabel('Months', fontsize=14)
plt.ylabel('Number Of Cancellations', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#prepare data for seasonal analysis
cancellations_per_month = data_set.groupby('arrival_date_month')['is_canceled'].sum()
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
cancellations_per_month = cancellations_per_month.reindex(months_order)

#line chart for cancellations over time
plt.figure(figsize=(12, 6))
cancellations_per_month.plot(kind='line', marker='o')
plt.title('Cancellations Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Cancellations')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

Seasonal trends indicate a spike in cancellations during March, April, and July, which may correlate with spring break, summer vacations, or other seasonal events.

This pattern can help in anticipating high-risk periods and tailoring strategies to minimize cancellations. You could consider offering special packages, flexible booking options, or additional incentives during these peak months to encourage customers to stick with their reservations.

# Special Requests:

- Customers with no special requests tend to cancel more often. This could indicate that customers who make special requests are more committed to their bookings, possibly because their travel plans are more personalized.

These factors suggest that managing cancellation risks could involve targeting customers with long lead times and reducing cancellations through strategic communication and offers during peak seasons.
